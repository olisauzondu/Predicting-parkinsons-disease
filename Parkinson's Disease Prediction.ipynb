#### Feature selection

```python
#rearrange columns for feature selection
parkdf.columns
```

```python
parkdf = parkdf[['name', 'MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE', 'status']]

#create copy of data
df = parkdf.copy()

#convert categorical columns to numeric (only one object type column - name)
num = LabelEncoder()
df['name'] = num.fit_transform(df['name'])
df.head()
```

```python
X = df.iloc[:,0:14] #features
Y = df.iloc[:,-1] #target variable. parkinsons or not

best_features = SelectKBest(score_func=chi2,k=5) #function that select top 5 features
fit = best_features.fit(X,Y)
# creat dataframe for features and score of features
parkinson_scores = pd.DataFrame(fit.scores_)
parkinson_columns = pd.DataFrame(X.columns)
```

```python
#create a dataframe that combine all features and their scores
features_scores = pd.concat([parkinson_columns,parkinson_scores],axis=1)
features_scores.columns = ['Feature','Score']
features_scores.sort_values(by='Score')
```

It seems the top features are:
 - MDVP:Flo(Hz)
 - MDVP:Fo(Hz)
 - MDVP:Fhi(Hz)

The others don't have a high score as those three

#### Apply techniques
Build model

```python
x = parkdf[['MDVP:Flo(Hz)','MDVP:Fo(Hz)','MDVP:Fhi(Hz)']]
y = parkdf['status']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=7)
model = XGBClassifier() #xgbclassifier used for categorical target (classification problem) 
model.fit(x_train,y_train)
```

Evaluate model

```python
y_pred = model.predict(x_test)
accuracy_score(y_test,y_pred)
```

The model has an **89.8% accuracy** in prediction

```python
#define metrics
y_pred_prob = model.predict_proba(x_test) [:,1]

#calulate true and false positive rates
false_positive_rate,true_positive_rate,_ = metrics.roc_curve(y_test,y_pred_prob)

#calculate area under roc curve
auc = metrics.roc_auc_score(y_test,y_pred_prob)

#Create ROC curve
plt.plot(false_positive_rate, true_positive_rate,label="AUC="+str(auc))
plt.title('ROC Curve')
plt.ylabel('True Positive Rate')
plt.xlabel('false Positive Rate')
plt.legend(loc='lower right')
```

With the **area under curve of about 0.87**, means the model is performing well enough.

Save model for later use

```python
import pickle
park_model = 'park_model.sav'
pickle.dump(model,open(park_model, 'wb'))
```

```python
#to make predictions, input format to code: (MDVP:Flo(Hz),MDVP:Fo(Hz),MDVP:Fhi(Hz))
x_pred = np.array([[150,174,118]])
#load model for use using
loaded_model = pickle.load(open(park_model,'rb'))
result=loaded_model.predict(x_pred)
print(result)
```
